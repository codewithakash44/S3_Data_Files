```python
import pandas as pd
import random

# Define column names
columns = [
    "ID", "Year_Birth", "Education", "Marital_Status", "Income", 
    "Kidhome", "Teenhome", "Dt_Customer", "Recency", 
    "MntWines", "MntFruits", "MntMeatProducts", "MntFishProducts", 
    "MntSweetProducts", "MntGoldProds", "NumDealsPurchases", 
    "AcceptedCmp1", "AcceptedCmp2", "AcceptedCmp3", "AcceptedCmp4", 
    "AcceptedCmp5", "Response", "NumWebPurchases", "NumCatalogPurchases", 
    "NumStorePurchases", "NumWebVisitsMonth"
]

# Define a base row
base_row = [
    1, 1985, "Graduation", "Single", 58000, 
    0, 1, "2012-04-01", 10, 
    300, 50, 240, 20, 30, 10, 5, 
    1, 0, 0, 1, 
    0, 6, 2, 5, 
    3, 5
]

# Generate data with duplicates and variations
data = []
for i in range(40):
    if i % 5 == 0:  # Add duplicate rows every 5 rows
        data.append(base_row)
    else:
        # Create variations by modifying some values
        row = base_row.copy()
        row[0] = i + 1  # Unique ID
        row[1] = random.choice(range(1970, 2000))  # Random Year_Birth
        row[3] = random.choice(["Single", "Married", "Divorced"])  # Random Marital_Status
        row[4] = random.choice(range(40000, 100000, 5000))  # Random Income
        row[8] = random.randint(1, 30)  # Random Recency
        row[9] = random.randint(100, 600)  # Random MntWines
        row[-1] = random.randint(1, 10)  # Random NumWebVisitsMonth
        data.append(row)

# Load data into a DataFrame
df = pd.DataFrame(data, columns=columns)

# Display the DataFrame
print(df)

# Optional: Save to a CSV file for testing
df.to_csv("customer_data.csv", index=False)
```

```python
#3.	Create Aggregate Features:
        o	Generate a new column, Total Products, by summing up product-related columns (MntWines, MntFruits, MntMeatProducts, MntFishProducts, MntSweetProducts, MntGoldProds).
        o	Create Total Purchases by summing up purchase-related columns (NumDealsPurchases, NumWebPurchases, NumCatalogPurchases, NumStorePurchases).
        o	Calculate Total Campaigns by summing the campaign-related columns (AcceptedCmp1 through AcceptedCmp5).
        o	Combine Kidhome and Teenhome into a single column, Total Home, representing the total number of children or teens at home.
```

```python
Task 4: Model Building and Data Splitting
Steps:
1.	Define Features and Target Variable:
        o	Select the feature set (X) by including the first five columns.
        o	Set the target variable (Y) as the last column of the DataFrame.
2.	Split the Data into Training and Testing Sets:
        o	Divide the data into training and testing sets with an 80-20 split.
        o	Use a random state of 42 and stratify the target variable (Y) using the train_test_split function.
```

```python
#1.	Define Features and Target Variable:
#        o	Select the feature set (X) by including the first five columns.
#        o	Set the target variable (Y) as the last column of the DataFrame.  

# Selecting the first five columns as features (X)
X = df.iloc[:, :5]

# Setting the target variable (Y) as the last column
Y = df.iloc[:, -1]
```

```python
#2.	Split the Data into Training and Testing Sets:
#        o	Divide the data into training and testing sets with an 80-20 split.
#        o	Use a random state of 42 and stratify the target variable (Y) using the train_test_split function.

from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42, stratify=Y)

```

```python
Task 5: Handle Class Imbalance with Resampling Techniques
Steps:
1.	Apply SMOTE to Oversample the Minority Class:
        o	Use SMOTE with a random state of 42 to generate synthetic samples for the minority class, helping to balance the dataset.
        o	Apply SMOTE on X_train and Y_train to create resampled training data (X_train_res, Y_train_res).
2.	Apply Random Undersampling for the Majority Class:
        o	Initialize RandomUnderSampler with a random state of 42 to reduce samples from the majority class.
        o	Apply undersampling on the already resampled data (X_train_res and Y_train_res).
3.	Combine SMOTE and ENN (Edited Nearest Neighbors) for Enhanced Resampling:
        o	Use the SMOTEENN technique with a random state of 42 to apply both oversampling and undersampling, balancing classes and removing noisy samples.
        o	Resample X_train and Y_train using SMOTEENN to produce a final balanced training dataset (X_train_res, Y_train_res).
```

```python
print(Y_train.value_counts())

import numpy as np

# Get the indices of the Y_train values
indices = Y_train.index

# Randomly select a few indices to modify
num_to_replace = 14  # Number of values to replace
random_indices = np.random.choice(indices, size=num_to_replace, replace=False)

# Replace the selected indices with a different class value
Y_train.loc[random_indices] = 0.0

# Check the updated class distribution
print(Y_train.value_counts())
```

```python
#1.	Apply SMOTE to Oversample the Minority Class:
#        o	Use SMOTE with a random state of 42 to generate synthetic samples for the minority class, helping to balance the dataset.
#        o	Apply SMOTE on X_train and Y_train to create resampled training data (X_train_res, Y_train_res).

from imblearn.over_sampling import SMOTE

# Initialize SMOTE with a random state of 42
smote = SMOTE(random_state=42)

# Apply SMOTE on the training data to balance the classes
X_train_res, Y_train_res = smote.fit_resample(X_train, Y_train)
```

```python
#2.	Apply Random Undersampling for the Majority Class:
#    o	Initialize RandomUnderSampler with a random state of 42 to reduce samples from the majority class.
#    o	Apply undersampling on the already resampled data (X_train_res and Y_train_res).

from imblearn.under_sampling import RandomUnderSampler

# Initialize RandomUnderSampler with a random state of 42
undersampler = RandomUnderSampler(random_state=42)

# Apply undersampling on the resampled data
X_train_res, Y_train_res = undersampler.fit_resample(X_train_res, Y_train_res)

```

```python
#3.	Combine SMOTE and ENN (Edited Nearest Neighbors) for Enhanced Resampling:
#    o	Use the SMOTEENN technique with a random state of 42 to apply both oversampling and undersampling, balancing classes and removing noisy samples.
#    o	Resample X_train and Y_train using SMOTEENN to produce a final balanced training dataset (X_train_res, Y_train_res).

from imblearn.combine import SMOTEENN

# Initialize SMOTEENN with a random state of 42
smote_enn = SMOTEENN(random_state=42)

# Apply SMOTEENN on the training data
X_train_res, Y_train_res = smote_enn.fit_resample(X_train, Y_train)

# Check the distribution of the resampled target variable
print(Y_train_res.value_counts())
```

```python
Task 6: Configure and Train the k-Nearest Neighbors (k-NN) Model
Steps:
1.	Determine the Optimal k Value:
    o	Calculate the initial value of k using the square root of the total length of data in the dataset (sqrt(n)).
    o	Adjust k to an odd value (e.g., 47) for better distance calculations and prediction consistency.
2.	Initialize and Train the k-NN Model:
    o	Set up a KNeighborsClassifier with n_neighbors=k (the determined optimal k). Store the instance in variable model.
    o	Train (fit) the k-NN model on the resampled training dataset (X_train_res, Y_train_res) to prepare for predictions.
3.	Predict on the Test Set:
    o	Use the trained k-NN model to make predictions on X_test, storing the predicted values in y_pred.
```

```python
#1.	Determine the Optimal k Value:
#    o	Calculate the initial value of k using the square root of the total length of data in the dataset (sqrt(n)).
#    o	Adjust k to an odd value (e.g., 47) for better distance calculations and prediction consistency.

import math

# Calculate the total number of samples in the dataset
n = len(X_train_res)  # Using the resampled training dataset

# Calculate the initial value of k as the square root of n
k = int(math.sqrt(n))

# Adjust k to the nearest odd value for better distance calculations
if k % 2 == 0:
    k += 1  # Make k odd if it's even

# Print the calculated value of k
print(f"The optimal value of k is: {k}")
```

```python
#2.	Initialize and Train the k-NN Model:
#    o	Set up a KNeighborsClassifier with n_neighbors=k (the determined optimal k). Store the instance in variable model.
#    o	Train (fit) the k-NN model on the resampled training dataset (X_train_res, Y_train_res) to prepare for predictions.
#3.	Predict on the Test Set:
#    o	Use the trained k-NN model to make predictions on X_test, storing the predicted values in y_pred.

from sklearn.neighbors import KNeighborsClassifier

# Initialize the k-NN model with the determined optimal k
model = KNeighborsClassifier(n_neighbors=k)

# Train (fit) the k-NN model on the resampled training dataset
model.fit(X_train_res, Y_train_res)

# Predict on the test set
y_pred = model.predict(X_test)

# Output predictions
print("Predictions on the test set:")
print(y_pred)
```

```python
Task 7: Evaluate Model Performance Using Confusion Matrix and Classification Report
Steps:
1.	Generate the Confusion Matrix:
        o	Use the confusion_matrix function under sklearn to create a confusion matrix comparing predicted values (y_pred) to actual test labels (Y_test).
        o	Display the confusion matrix to evaluate the accuracy of the model's predictions across each class.
2.	Create the Classification Report:
        o	Generate a classification report using classification_report with the parameters (y_pred and Y_test).
        o	Set output_dict=True to convert the report into a dictionary format for easier data manipulation. Store it in variable cr.
3.	Visualize the Confusion Matrix:
        o	Use a heatmap to visualize the confusion matrix for a more intuitive assessment of model performance, applying the coolwarm color map for clarity.
```

```python
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Generate the Confusion Matrix
cm = confusion_matrix(Y_test, y_pred)  # Compare actual test labels (Y_test) with predicted values (y_pred)

# Display the confusion matrix
print("Confusion Matrix:")
print(cm)

# Step 2: Create the Classification Report
cr = classification_report(Y_test, y_pred, output_dict=True)  # Generate the classification report in dictionary format
print("\nClassification Report:")
print(cr)

# Step 3: Visualize the Confusion Matrix
plt.figure(figsize=(10, 8))  # Set the size of the heatmap visualization
sns.heatmap(cm, annot=True, fmt='d', cmap='coolwarm', xticklabels=np.unique(Y_test), yticklabels=np.unique(Y_test))
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix Heatmap")
plt.show()
```
